```python
# using converter
def rename(name):
    if name == "Male":
        return 1
    elif name == "Female":
        return 0
    else:
        return -1

df = pd.read_csv(r'C:\Users\Laptop Valley\Desktop\AI\Machine_learning\data_Preprcossing\loan_data_set.csv' , converters={'Gender': rename,})
df
```

```python
# converting csv into json
df.to_json(r'C:\Users\Laptop Valley\Desktop\AI\Machine_learning\data_Preprcossing\loan_data_set.json', orient='records')
# what is orient: records
# It means that each record (row) will be a separate JSON object in the output file.
# why it is useful?
# This format is useful for APIs and data interchange between systems, as it allows for easy parsing and manipulation of individual records.
```
