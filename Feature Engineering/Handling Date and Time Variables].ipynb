```python
import pandas as pd 
import numpy as np 
df = pd.read_csv("orders.csv")
df1 = pd.read_csv("messages.csv")
df.head()
```

```python
df["date"] =pd.to_datetime(df["date"])
df.info()
```

```python
# Extract year
df["year"]=df["date"].dt.year
df.head()
```

```python
# Extract month
df["month"]=df["date"].dt.month_name()
df.head()
```

```python
# Extract day
df["day"]=df["date"].dt.day
df.head()
```

```python
# Extract day of week
df["day"]=df["date"].dt.day_name()
df.head()
```

```python
# Extract day of week
df["day"]=df["date"].dt.day_of()
df.head()
```
