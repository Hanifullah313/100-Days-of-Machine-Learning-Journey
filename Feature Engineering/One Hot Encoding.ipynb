```python
import pandas as pd 
import numpy as np 
df = pd.read_csv("cars.csv" , header=0)
df.head()
```

```python
df["fuel"].value_counts()
```

# One Hot Encoding Using Pandas

```python
pd.get_dummies(df , columns= ["fuel" ,"owner"])
```

# K-1 One Hot Encoding Using Pandas

```python
pd.get_dummies(df , columns = ["fuel" ,"owner"] , drop_first = True)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1], test_size = 0.3 , random_state = 2)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop="first")
x_train_ohe=ohe.fit_transform(x_train[["fuel" , "owner"]] ).toarray()
x_test_ohe=ohe.fit_transform(x_test[["fuel" , "owner"]] ).toarray()
x_train_ohe
```
