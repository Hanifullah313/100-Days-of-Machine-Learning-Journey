```python

class SGDRegressor:
    
    def __init__(self, learning_rate=0.01, epochs=100):
        # Model parameters
        self.coef_ = None        # coefficients (w1, w2, ..., wn)
        self.intercept_ = None   # intercept (b)
        self.lr = learning_rate  # learning rate (Î±)
        self.epochs = epochs     # number of passes over dataset
        
    def fit(self, X_train, y_train):
        # Initialize intercept and coefficients
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])  # start with wj = 1
        
        # Run for the given number of epochs
        for i in range(self.epochs):
            for j in range(X_train.shape[0]):
                
                # Randomly pick one sample (stochastic)
                idx = np.random.randint(0, X_train.shape[0])
                
                # Prediction for that one sample: Å· = xiÂ·w + b
                y_hat = np.dot(X_train[idx], self.coef_) + self.intercept_
                
                # -------- Gradient for Intercept (bias term) --------
                # âˆ‚L/âˆ‚b = -2 * (y - Å·)
                intercept_der = -2 * (y_train[idx] - y_hat)
                
                # Update rule: b := b - Î± * âˆ‚L/âˆ‚b
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)

                # -------- Gradient for Coefficients (weights) --------
                # âˆ‚L/âˆ‚w = -2 * xi * (y - Å·)
                coef_der = -2 * np.dot((y_train[idx] - y_hat), X_train[idx])
                
                # Update rule: w := w - Î± * âˆ‚L/âˆ‚w
                self.coef_ = self.coef_ - (self.lr * coef_der)
        
        # Print learned parameters
        print("Final intercept (b):", self.intercept_)
        print("Final coefficients (w):", self.coef_)
    
    def predict(self, X_test):
        # Prediction formula: Å· = XÂ·w + b
        return np.dot(X_test, self.coef_) + self.intercept_
```
